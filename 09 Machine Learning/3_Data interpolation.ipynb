```python
# Data interpolation >> is a process pf estimating unknow values from the values

# linear interpolation
# cubic interplotion
# polynomial interpolation
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 3, 4, 5, 7,])

plt.scatter(x, y)
```

```python
# linear interplotation >> linear in nature
x_new = np.linspace(1, 5, 10) # create new x value
y_interp = np.interp(x_new, x, y)
```

```python
y_interp
```

```python
plt.scatter(x_new, y_interp)
```

```python
# cubeic interpolation
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 8, 27, 64, 125])
plt.scatter(x, y)
```

```python
from scipy.interpolate import interp1d
f = interp1d(x, y, kind = 'cubic')
```

```python
f
```

```python
x_new = np.linspace(1, 5, 10)
```

```python
y_interp = f(x_new)
y_interp
```

```python
plt.scatter(x_new, y_interp)
```

```python
# polynomial interplotion
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 1, 25])

p = np.polyfit(x, y, 2)
```

```python
x_new = np.linspace(x, y, 2)
y_interp = np.polyval(p, x_new)
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x_new, y_interp)
```

```python
x_new = np.linspace(x, y, 10)
y_interp = np.polyval(p, x_new)
```

```python
plt.scatter(x_new, y_interp)
```

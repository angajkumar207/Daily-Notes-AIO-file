##  Polynomial regression:
### -Polynomial regression is just linear regression with extra powers of the input variable(s).

### Example: Instead of fitting a straight line
  - y=b0​+b1​x

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Genrate same synthetics dp or repraduciblitiy of results
np.random.seed(1)
```

```python
# Genrate synthtic X
x = 2*np.random.rand(100, 1)
```

```python
x
```

```python
# Generate random sythetic y usiong x
y = 4 + 3*x+1.5*x**2 + np.random.rand(100, 1)
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,  y, test_size=0.25, random_state=1)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# perfrom polynomial regression
from sklearn.preprocessing import PolynomialFeatures
poly_features =  PolynomialFeatures(degree=2, include_bias=False)
```

```python
poly_features
```

```python
x_poly_train = poly_features.fit_transform(x_train)
```

```python
x_poly_train
```

```python
from sklearn.linear_model import LinearRegression
poly_reg = LinearRegression()
```

```python
poly_reg
```

```python
poly_reg.fit(x_poly_train, y_train)
```

```python
poly_reg.coef_
```

```python
poly_reg.intercept_
```

```python
# make prediction on train data itself for the evaluation metrics
y_poly_pred = poly_reg.predict(x_poly_train)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train, y_poly_pred)
```

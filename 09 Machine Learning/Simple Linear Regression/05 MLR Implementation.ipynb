```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes
```

```python
print(diabetes.DESCR)
```

```python
# Problem:
# To predict disease progression one year after the baseline.
```

```python
# data
diabetes.data
```

```python
diabetes.target
```

```python
# feature name
diabetes.feature_names
```

```python
# creae data frame
data = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
```

```python
data
```

```python
# Add target
data['target'] = diabetes.target
```

```python
data
```

```python
# EDA data preparation
x = data.drop('target', axis = 1)
y = data['target']
```

```python
x
```

```python
y
```

```python
# Split the data into train and test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

```

```python
# sape of x_train, x_test
x_train.shape, x_test.shape
```

```python
#scaling(optional)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler
```

```python
# scaling [HW]
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# Model Training 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train.shape
```

```python
model.coef_
```

```python
# check of lenth 
len(model.coef_)
```

```python
# check model intercept
model.intercept_
```

```python
# interpretation
# say age has coef value -1.44969259
# interpretation >> 1 unit change in age then target value decrease by -30 units on an average when all
# other features/factors are kept constant
x_train
data
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

## Simple linear Regression Python implementation - 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data from scikit learn
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes
```

```python
# Problem statement: To predict disease progression one year after baseline provide 10 features
```

```python
print(diabetes.DESCR)
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
diabetes.feature_names
```

```python
# Dataframe
data = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
# add new columns 
data['target'] = diabetes.target
```

```python
data
```

```python
x = data[['bmi']] # one independent variable
y = data[['target']] # dependent variable
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=1)
# random_state when fixed helps us to reproduce the same result again again again
```

```python
x_train
```

```python
x_test
```

```python
# total row x_trn and x_tst
353+89
```

```python
y_train
```

```python
# scaling(optional)
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
x_test
```

```python
x_train
```

```python
# training of model
model.fit(x_train, y_train)
# at the backend it will use gradient descent to give optional coefficient
```

```python
model.coef_ # atteribute of the model
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# visualise the result
plt.scatter(x_test, y_test, color = "black", label = "Actual")
plt.plot(x_test, y_test, color = "blue", linewidth = 3, label = "Predicted linear regression line ")
plt.xlabel("BMI")
plt.ylabel("One year progression target")
plt.title("Linear regression on diabetes data ")
plt.legend()
plt.show
```

```python
plt.plot(x_test, y_test, color = "blue", linewidth = 3, label = "Predicted linear regression line ")
plt.xlabel("BMI")
plt.ylabel("One year progression target")
plt.title("Linear regression on diabetes data ")
plt.legend()
plt.show
```

```python
# model evaluation
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

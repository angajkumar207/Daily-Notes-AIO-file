```python
from sklearn.preprocessing import OneHotEncoder
df = pd.DataFrame({"status": ["single", "married", "separeted", "single", "married"]})
```

```python
df
```

```python
encoder = OneHotEncoder()
```

```python
encoder
```

```python
encoded = encoder.fit_transform(df[["status"]]).toarray()
```

```python
encoded 
```

```python
encoder.get_feature_names_out()
```

```python
encoder_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
encoder.transform([["single"]]).toarray()
```

```python
from sklearn.preprocessing import OrdinalEncoder
df = pd.DataFrame({"qualification": ["HS", "CR", "PG", "PHD", "BCA"]})
```

```python
df
```

```python
encoder = OrdinalEncoder(categories=[["HS", "CR", "PG", "PHD", "BCA"]])
```

```python
encoder.fit_transform(df[["qualification"]])
```

```python
# New Data 
encoder.transform([["PG"]])
```

```python
import pandas as pd

df = pd.DataFrame({
    'time': ["lunch", "breakfast", "dinner", "lunch", "breakfast", "lunch", "dinner", "lunch", "dinner"],
    'total_bill': [120, 130, 90, 125, 150, 190, 160, 180, 189]
})
```

```python
df
```

```python
mean_price = df.groupby('time')['total_bill'].mean()
```

```python
df["timne_encoder"] = df['time'].map(mean_price)
```

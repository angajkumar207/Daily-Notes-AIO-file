```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/8217ecc82f6d11e8aac357d3aa5a9f362663c388/titanic.csv')  # à¤¯à¤¾ seaborn à¤¸à¥‡ à¤²à¤¿à¤¯à¤¾ à¤—à¤¯à¤¾ Titanic dataset

male_names = ["John", "Robert", "Michael", "William", "David"]
female_names = ["Mary", "Patricia", "Linda", "Barbara", "Elizabeth"]


def assign_name(sex):
    if sex == 'male':
        return np.random.choice(male_names)
    elif sex == 'female':
        return np.random.choice(female_names)
    else:
        return "Unknown"


df['name'] = df['sex'].apply(assign_name)


print(df[['sex', 'name']].head())
```

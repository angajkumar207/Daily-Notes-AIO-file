```python
df3 = pd.DataFrame({
    'key1': [1, 2, 3, 4],
    'key2': [11, 21, 31, 41],
    'key3': [41, 42, 43, 45]
}
)
```

```python
df4 = pd.DataFrame({
    'key1': [1, 2, 5, 6],
    'key4': [11, 21, 31, 41],
    'key5': [541, 542, 543, 545]
}
)
```

```python
pd.merge(df3, df4, how = 'inner') #print same rows element 
```

```python
pd.merge(df3, df4, how = 'left')
```

```python
pd.merge(df3, df4, how = 'outer') # all the common element of the common columns from both the Data frame
```

```python
pd.merge(df3, df4, how = 'cross')
```

```python
pd.merge(df3, df4, how = 'left', left_on="key2", right_on="key4") # all elemet same key2 and key4 so no change Nan.
```

```python
 # FUNCTION
def create_flag(x):
    if x < 10:
        return "cheap"
    if x >= 10 and x <= 20:
        return "medium"
    else:
        return "High"
```

```python
df['flag_fare'] = df['fare'].apply(create_flag)
```

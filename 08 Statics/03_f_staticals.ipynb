```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Normal disturbution >> gaussion disturibution of data point is clustered around mean 
mu = 0
sigma = 0.1

s = np.random.normal(mu, sigma, size = 10000)
```

```python
len(s)
```

```python
plt.hist(s, 30)
plt.show()
```

```python
sns.distplot(s)
```

```python
# uniform disturbution >> all the outcomes has equal probability to occur in a range
u = np.random.uniform(0, 1, 10000)
```

```python

plt.hist(u)
plt.show()
```

```python
sns.distplot(u)
```

```python
# long normal distribution
l = np.random.lognormal(mean = 0, sigma=0.5, size = 1000)
plt.hist(1, bins = 100)
plt.show()
```

```python
sns.boxplot(l) # since lognormal is right skewed, so outliers are  
```

```python
sns.distplot(l)
```

```python
# lognormal >> normal disturbution
```

```python
sns.distplot(np.log(l))
```

```python
sns.distplot(np.sqrt(l))
```

```python
# chi square disturibution
# multiple chi square dist plot
c = np.random.chisquare(df = 30, size = 10000)
d = np.random.chisquare(df = 10, size = 10000)
c = c[np.where(abs(c) < 100)]
d = d[np.where(abs(d) < 100)]
```

```python
plt.hist(c, bins = 1000)
plt.hist(d, bins = 1000)
plt.show()
```

```python
# T disturibution 
t = np.random.standard_t(df=3, size = 10000)
t = t[np.where(abs(t) < 100)]
plt.hist(t, bins = 1000)
plt.show()
```

```python
t = np.random.standard_t(df=4, size = 10000)
t = t[np.where(abs(t) < 100)]
plt.hist(t, bins = 1000)
plt.show()
```

```python
# f disturibution 
f = np.random.f(dfnum = 100, dfden = 100, size = 10000)
f = f[np.where((abs(f)) < 10)]
plt.hist(f, bins = 10000)
plt.show()
```

```python
# discreate distribution 
# bernoully dist >> binomial dist with one trial is bernoully dist

b = np.random.binomial(n=1, p = 0.5, size= 1000)
plt.hist(b)
plt.show()
```

```python
# binomial dist
b = np.random.binomial(n=100, p = 0.5, size= 1000)
plt.hist(b)
plt.show()
```

```python
# poission distribution
p = np.random.poisson(lam = 5, size = 100)
plt.hist(p, bins=20)
plt.show()
```
